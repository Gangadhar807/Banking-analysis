```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Banking.csv")

bins = [0,100000,300000, float('inf')]
tabels = ['Low', 'Med', 'High']
df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins, labels=tabels, right=False)

for i, predictor in enumerate(df[['BRId', 'GenderId', 'IAId', 'Amount of Credit Cards', 'Nationality', 'Occupation', 'Fee Structure', 'Loyalty Classification', 'Properties Owned', 'Risk Weighting', 'Income Band']]):
  plt.figure(i)
  sns.countplot(data=df, x=predictor)
```

```python
for i, predictor in enumerate(df[['BRId', 'GenderId', 'IAId', 'Amount of Credit Cards', 'Nationality', 'Occupation', 'Fee Structure', 'Loyalty Classification', 'Properties Owned', 'Risk Weighting', 'Income Band']]):
  plt.figure(i)
  sns.countplot(data=df, x=predictor, hue='GenderId')
```

```python
for i, predictor in enumerate(df[['BRId', 'GenderId', 'IAId', 'Amount of Credit Cards', 'Nationality', 'Occupation', 'Fee Structure', 'Loyalty Classification', 'Properties Owned', 'Risk Weighting', 'Income Band']]):
  plt.figure(i)
  sns.countplot(data=df, x=predictor, hue='Nationality')
```

```python
# Histoplot of values counts for diffrent Ocuption

for col in categorical_cols:
  if col == "Occupation":
    continue
  plt.figure(figsize=(8,4))
  sns.histplot(df[col])
  plt.title('Histogram of Occuption Count')
  plt.xlabel(col)
  plt.ylabel('Count')
  plt.show()
```

numarical analysis

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']

# Univariate analysis and visualization
plt.figure(figsize=(8,4)) # Increased figure width to accommodate 3 plots
for i, col in enumerate(numerical_cols):
  plt.subplot(3,4,i+1) # Changed from 1,2 to 1,3 to fit all 3 plots
  sns.histplot(df[col], kde=True)
  plt.title(col)
plt.tight_layout() # Add tight_layout to prevent labels from overlapping
plt.show()
```

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']

correation_matrix = df[numerical_cols].corr()
plt.figure(figsize=(8,4))
sns.heatmap(correation_matrix, annot=True, cmap='crest',fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
